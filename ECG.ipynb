```python
#REF:https://physionet.org/content/mitdb/1.0.0/
#http://www.biomedres.info/biomedical-research/ecg-signal-classification-and-parameter-estimation-using-multiwavelet-transform.html#11
#https://physionet.org/lightwave/?db=mitdb/1.0.0


import wfdb 
import matplotlib.pyplot as plt
import numpy as np

record = wfdb.rdrecord('100', sampto=3000)
annotation = wfdb.rdann('100', 'atr', sampto=3000)

wfdb.plot_wfdb(record=record, annotation=annotation, plot_sym=True, time_units='seconds', title='MIT-BIH Record 100', figsize=(10,4))
```

```python
sig, fields = wfdb.rdsamp('100', channels=[0])
```

```python
sig
```

```python
fields
```

```python
plt.plot(sig[0:1000])
plt.grid(color='b', linestyle='-', linewidth=0.5)
```

```python
from scipy import signal
```

```python
sigforspec = sig.flatten()
```

```python
np.shape(sigforspec)
```

```python
from matplotlib import pyplot as plt
freqs, times, spectrogram = signal.spectrogram(sigforspec[0:1000])
```

```python
plt.imshow(spectrogram, aspect='auto', cmap='hot_r', origin='lower')
plt.ylabel('Frequency band')
plt.xlabel('Time window')
plt.ylim((0,20))
```

```python
freqs, psd = signal.welch(sigforspec)
```

```python
plt.semilogx(freqs, psd)
plt.xlabel('Frequency')
plt.ylabel('Power')
```
